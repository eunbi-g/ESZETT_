```python
data1 = pd.read_csv('LOCAL_PEOPLE_GU_201801.csv',index_col=False, engine='python',encoding='cp949')
data2 = pd.read_csv('LOCAL_PEOPLE_GU_201802.csv',index_col=False, engine='python',encoding='cp949')
data3 = pd.read_csv('LOCAL_PEOPLE_GU_201803.csv',index_col=False, engine='python',encoding='cp949')
data4 = pd.read_csv('LOCAL_PEOPLE_GU_201804.csv',index_col=False, engine='python',encoding='cp949')
data5 = pd.read_csv('LOCAL_PEOPLE_GU_201805.csv',index_col=False, engine='python',encoding='cp949')
data6 = pd.read_csv('LOCAL_PEOPLE_GU_201806.csv',index_col=False, engine='python',encoding='cp949')
data7 = pd.read_csv('LOCAL_PEOPLE_GU_201807.csv',index_col=False, engine='python',encoding='cp949')
data8 = pd.read_csv('LOCAL_PEOPLE_GU_201808.csv',index_col=False, engine='python',encoding='cp949')
data9 = pd.read_csv('LOCAL_PEOPLE_GU_201809.csv',index_col=False, engine='python',encoding='cp949')
data10 = pd.read_csv('LOCAL_PEOPLE_GU_201810.csv',index_col=False, engine='python',encoding='cp949')
data11 = pd.read_csv('LOCAL_PEOPLE_GU_201811.csv',index_col=False, engine='python',encoding='cp949')
data12 = pd.read_csv('LOCAL_PEOPLE_GU_201812.csv',index_col=False, engine='python',encoding='cp949')
```

```python
# 2018 01 - 2018 12 데이터 병합
df18 = pd.concat([data1,data2,data3,data4,data5,data6,data7,data8,data9,data10,data11,data12])
df = pd.concat([df17,df18])
```

```python
#column명 변경
for i in range(1,12,2):
    df['m'+str(i//2+1)+'0'] = df.iloc[:,i+4:i+6].sum(axis=1)
    df['f'+str(i//2+1)+'0'] = df.iloc[:,i+18:i+21].sum(axis=1)
df.head()
```

```python
#cloumn명 영문 변환
df = df.rename(columns = {'기준일ID':'date'})
df = df.rename(columns = {'시간대구분':'time'})
df = df.rename(columns = {'자치구코드':'local code'})
df = df.rename(columns = {'총생활인구수':'total pop'})

df = df.rename(columns = {'남자0세부터9세생활인구수':'m0'})
df = df.rename(columns = {'남자70세이상생활인구수':'m70'})
df = df.rename(columns = {'여자0세부터9세생활인구수':'f0'})
df = df.rename(columns = {'여자70세이상생활인구수':'f70'})
```

```python
df = df.drop(df.iloc[:,5:17], axis='columns')
```

```python
df = df.drop(df.iloc[:,7:19], axis='columns')
df.head()
```

```python
# 서울시 자치구 코드 파일
GU_Map = pd.read_csv('WGS.csv', index_col=False)
```

```python
del GU_Map['순번']
del GU_Map['ESRI_PK']

GU = GU_Map.sort_values(by='시군구코드')
GU
```

```python
GU = GU.rename(columns = {'시군구코드':'local code'})
GU = GU.rename(columns = {'시군구명_한글':'local_kor'})
GU = GU.rename(columns = {'시군구명_영문':'local_eng'})
GU = GU.rename(columns = {'위도':'latitude'})
GU = GU.rename(columns = {'경도':'longitude'})
GU
```

```python
final = pd.merge(df,GU)
final
```

```python
final.to_csv('local_people_gu_17_18.csv', index=False, encoding='cp949')
```
